RANDOM NUMBERS
- OFTEN USED TO INCORPORATE A RANDOM VALUE IN YOUR PROGROMMAMING

MID SQUARED METHOD
- STARTS WITH A SEED VALUE THAT IS SQUARED
- FROM THAT SQUARED VALUE, IT TAKES THE MIDDLE NUMBERS AND PRODUCES A NEW SEED
- IT IS OUTPUTTED, USUALLY USES THE TIME STAMP IN USERS COMPUTER AS THE SEED VALUE.

IMPORT RANDOM
- IMPORT THIS AND ASKS PYTHON TO GENERATE RANDOM NUMBERS

random.seed(x)
- sets the initial seed for the pseudorandom number generator
- if not specified it uses the current time stamp as the seed value

random.random()
- returns a pseudorandom float between 0.0 and 1.0

random.randrange (start, stop) or (1,10)
- returns a psuedo random int between start and stop. ***Stop value excluded***

random.randint(start,stop) or (1,10)
- returns between start and stop both ***inclusive***

```python
import csv
import random

with open ("test.csv", "w") as file:
  writer = csv.writer(file, lineterminator ='\n')

  tail = 0
  head = 1

  for i in range(1, 101, 1):
    flip = random.randrange (1,3)

    if flip == 1:
      tail+= 1
      output = [i, "Tails",tail,head]
      writer.writerow(output)
    else:
      head += 1
      output = [i, "Heads",tail,head]
      writer.writerow(output)
```
